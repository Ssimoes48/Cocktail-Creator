```python
import pandas as pd
import numpy as np
from fractions import Fraction


csv_path = "Data/mr_boston_flattened.csv"
```

```python
category_df = pd.read_csv(csv_path, encoding="utf-8")
category_df.head()
```

```python
# drop columns from dataframe
type2_df = type_df = category_df[["name", "category"]]
type3_df = type2_df.rename(columns={'category':'type'})
type3_df.head()
```

## load to table

```python
from sqlalchemy import create_engine
from config import postgres_pwd, postgres_user

db_name = "cocktail_db"
table_name = "cocktail"

connection_string = f"{postgres_user}:{postgres_pwd}@localhost:5432/{db_name}"
engine = create_engine(f'postgresql://{connection_string}')
conn = engine.connect()

type3_df.to_sql(table_name, con=engine, if_exists="append", index=False)

conn.close()
```
