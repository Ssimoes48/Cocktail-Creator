```python
import pandas as pd
from fractions import Fraction

csv_path = "Data/mr_boston_flattened.csv"
```

```python
cocktail_df = pd.read_csv(csv_path, encoding="utf-8")
cocktail_df.head()
```

```python
recipe_df = cocktail_df[["name","glass","glass-size","instructions"]]
recipe_df
```

```python
recipe_df.to_csv("Data/recipe_dataframe.csv", encoding="utf-8", index=False, header=True)
```

```python
recipe2_df = recipe_df.rename(columns={'name':'cocktail', 'glass':'glass_type', 'glass-size':'glass_size'})
```

## Postgres for recipe table

```python
from sqlalchemy import create_engine
from config import postgres_pwd, postgres_user

db_name = "cocktail_db"
table_name = "recipe"

connection_string = f"{postgres_user}:{postgres_pwd}@localhost:5432/{db_name}"
engine = create_engine(f'postgresql://{connection_string}')
conn = engine.connect()

recipe2_df.to_sql(table_name, con=engine, if_exists="append", index=False)

conn.close()
```
