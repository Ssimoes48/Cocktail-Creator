```python
import pandas as pd
```

```python
filepath1 = "Data/statelatlong.csv"
state_df = pd.read_csv(filepath1)
state_df.head()
```

```python
filepath2 = "Data/state_cocktail_matched.csv"
cocktail_df = pd.read_csv(filepath2)
cocktail_df.head()
```

```python
combined_df = pd.merge(state_df, cocktail_df, on="state")
combined_df
```

```python
combined_df.to_csv("Data/combined_state.csv", encoding="utf-8", index=False, header=True)
final_df = combined_df.rename(columns={' cocktail':'cocktail', 'Image URL':'image_src'})
final_df
```

## Postgres for state table

```python
from sqlalchemy import create_engine
from config import postgres_pwd, postgres_user

db_name = "cocktail_db"
table_name = "state"

connection_string = f"{postgres_user}:{postgres_pwd}@localhost:5432/{db_name}"
engine = create_engine(f'postgresql://{connection_string}')
conn = engine.connect()

final_df.to_sql(table_name, con=engine, if_exists="append", index=False)

conn.close()
```
