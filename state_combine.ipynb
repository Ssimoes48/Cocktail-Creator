```python
import pandas as pd
```

```python
filepath1 = "Cocktail-Creator/Data/statelatlong.csv"

state_df = pd.read_csv(filepath1)


state_df.head()
```

```python
combined_df = pd.merge(state_df, cocktail_df, on="state")

combined_df
```

```python
combined_df.to_csv("Cocktail-Creator/Data/combined_state.csv", encoding="utf-8", index=False, header=True)
```
